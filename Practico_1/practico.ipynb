```python
import csv
import random


def generar_datos_controlados(n):
    estaturas = []
    pesos = []
    
    while len(estaturas) < n:
        estatura = round(random.uniform(1.30, 2.11), 2)
        peso = round(100 * (estatura - 1.30) + random.uniform(-10, 10), 2)
        
      
        if 1.30 <= estatura <= 2.11 and 40 <= peso <= 120:
            estaturas.append(estatura)
            pesos.append(peso)
    
    return estaturas, pesos


n = 100
estaturas, pesos = generar_datos_controlados(n)

with open('datos_generados.csv', 'w', newline='') as archivo:
    writer = csv.writer(archivo)
    writer.writerow(['Estatura', 'Peso'])
    for estatura, peso in zip(estaturas, pesos):
        writer.writerow([estatura, peso])

print("Datos generados y guardados en 'datos_generados.csv'.")


def leer_datos_csv(nombre_archivo):
    estaturas = []
    pesos = []
    try:
        with open(nombre_archivo, mode='r') as archivo:
            lector_csv = csv.reader(archivo)
            next(lector_csv)  
            for fila in lector_csv:
                estaturas.append(float(fila[0]))
                pesos.append(float(fila[1]))
    except FileNotFoundError:
        print(f"Error: El archivo '{nombre_archivo}' no se encuentra.")
    return estaturas, pesos

nombre_archivo = 'datos_generados.csv'
estaturas, pesos = leer_datos_csv(nombre_archivo)

if estaturas and pesos: 
    print("Estatura y Peso cargados:")
    for estatura, peso in zip(estaturas, pesos):
        print(f"Estatura: {estatura} m, Peso: {peso} kg")

    def calcular_regresion_lineal(x, y):
        n = len(x)
        suma_x = sum(x)
        suma_y = sum(y)
        suma_xy = sum(xi * yi for xi, yi in zip(x, y))
   
```
